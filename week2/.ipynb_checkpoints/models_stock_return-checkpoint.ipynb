### MODELS OF STOCK RETURN
---

### Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

### Reading data

```python
ms = pd.read_csv("../data/microsoft.csv", index_col=0)
ms.head(2)
```

### Distribution of return

```python
ms['Return'] = ms['Close'].shift(-1) - ms['Close']
ms["Return"] = ms["Return"]/ms["Close"] #transform to relative data

mu = ms["Return"].mean()
sigma = ms["Return"].std(ddof=1)

density = pd.DataFrame()
density["x"] = np.arange(ms["Return"].min() - 0.01, ms["Return"].max() + 0.01, 0.001)
density["pdf"] = norm.pdf(density["x"], mu, sigma)

plt.figure(figsize=(15,8))
ms["Return"].hist(bins=50)
plt.plot(density['x'], density['pdf'], color='red')
plt.title("Return histogram")
plt.xlabel("return")
plt.show()
```

```python
ms['LogReturn'] = np.log(ms['Close']).shift(-1) - np.log(ms['Close'])
mu = ms["LogReturn"].mean()
sigma = ms["LogReturn"].std(ddof=1)

density = pd.DataFrame()
density["x"] = np.arange(ms["LogReturn"].min() - 0.01, ms["LogReturn"].max() + 0.01, 0.001)
density["pdf"] = norm.pdf(density["x"], mu, sigma)

plt.figure(figsize=(15,8))
ms["LogReturn"].hist(bins=50)
plt.plot(density['x'], density['pdf'], color='red')
plt.title("LogReturn histogram")
plt.xlabel("LogReturn")
plt.show()
```

### Probability

```python
prob_return1 = norm.cdf(-0.05, mu, sigma) #drop over 5% per day
print("Probability (drop over 5%): {0}%".format(round(prob_return1*100, 4)))

prob_return2 = norm.cdf(-0.1, mu, sigma) #drop over 5% per day
print("Probability (drop over 10%): {0}%".format(prob_return2*100))
```

### Drop over year

```python
mu220 = 220*mu
sigma220 = (220**0.5)*sigma

prob_return3 = norm.cdf(-0.05, mu220, sigma220) #drop over 5% per year
print("Probability (drop over 5%): {0}%".format(round(prob_return3*100, 4)))

prob_return4 = norm.cdf(-0.40, mu220, sigma220) #drop over 40% per year
print("Probability (drop over 40%): {0}%".format(round(prob_return4*100, 4)))
```

### Calculate Value at risk (VaR)

```python
var = norm.ppf(0.05, mu, sigma) #with 5% chance, the daily return is worse than -2.25%
print("Single day value at risk: {0}".format(round(var, 4)))
```

```python
q25 = norm.ppf(0.25, mu, sigma)
print('25% quantile: ', round(q25, 4))

q75 = norm.ppf(0.75, mu, sigma) 
print('75% quantile: ', round(q75, 4))
```
