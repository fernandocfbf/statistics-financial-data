### Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Dice game

```python
die = pd.DataFrame([1,2,3,4,5,6])
results = [die.sample(2, replace=True).sum().loc[0] for i in range(50)]
```

### Frequency

```python
frequency = pd.DataFrame(results)[0].value_counts().sort_index()
frequency
```

```python
plt.figure(figsize=(10,8))
frequency.plot(kind="bar", color="blue")
plt.title("Frequency for sum in dice game")
plt.ylabel("Frequency")
plt.xlabel("Sum")
```

### Relative frequency

```python
frequency = pd.DataFrame(results)[0].value_counts(True).sort_index()

plt.figure(figsize=(10,8))
frequency.plot(kind="bar", color="blue")
plt.title("Frequency for sum in dice game")
plt.ylabel("Frequency")
plt.xlabel("Sum")
```

```python
results = [die.sample(2, replace=True).sum().loc[0] for i in range(10000)]
frequency = pd.DataFrame(results)[0].value_counts(True).sort_index()

plt.figure(figsize=(10,8))
frequency.plot(kind="bar", color="blue")
plt.title("Frequency for sum in dice game")
plt.ylabel("Frequency")
plt.xlabel("Sum")
```

### Expectation and variance

```python
x_distribution = pd.DataFrame({"Sum": [x for x in range(2,13)]}) #all possible sums
x_distribution["Probability"] = [1,2,3,4,5,6,5,4,3,2,1]
x_distribution["Probability"] = x_distribution["Probability"]/36
x_distribution
```

```python
mean = (x_distribution["Sum"]*x_distribution["Probability"]).sum()
var = ( ( ( x_distribution["Sum"]-mean )**2 )*x_distribution["Probability"] ).sum()
```

```python
print("Distribution mean: {0}\nDistribution variance {1}".format(mean, var))
```

### Empirical expectation and variance

```python
results = [die.sample(2, replace=True).sum().loc[0] for i in range(10000)]
results_formated = pd.Series(results)
```

```python
print("Empirical distribution mean: {0}\nEmpirical distribution variance {1}".format(results_formated.mean(), results_formated.var()))
```
